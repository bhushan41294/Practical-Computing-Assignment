```python
# Importing the libraries

import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Values of length, breath and height given in question

l = 30
b = 20
h = 15
```

```python
# Creating a function for finding area of one of the biggest face  

def biggest_face_area(l,b,h):
    face_1 = l*b
    face_2 = b*h
    face_3 = h*l
    return max([face_1,face_2,face_3])
```

```python
# Finding the biggest face area

biggest_face_area(30,20,15)
```

```python
# Creating a function for finding the volume of cuboid

def cuboid_volume(l,b,h):
    vol = l*b*h
    return vol
```

```python
# Finding the volume of cuboid by calling a function

cuboid_volume(30,20,15)
```

```python
# Predefined length in question  

length = [30,40,50]
for i in length:
    print(f"The biggest face area of cuboid with length {i} is {biggest_face_area(i,b,h)}")     
```

```python
# Finding the volume of cuboid wrt different lengths 

for i in length:
    print(f"The volume of cuboid with length {i} is {cuboid_volume(i,b,h)}")
```

### Non-vectorised method

```python
# Matrix multiplication 

mat_size = [10,30,50,100,500,1000]
non_vectorised_time = []
for n in mat_size:
    start = time.time() 
    matrix1 = np.random.randint(10,size = (n,n))
    matrix2 = np.random.randint(10,size = (n,n))
    res = np.zeros((n,n))
      
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                res[i][j] += matrix1[i][k] * matrix2[k][j]

    end = time.time()
    req_time = end - start
    non_vectorised_time.append(req_time)   
```

### Vectorised method

```python
# Matrix multiplication using Vectorised method

vectorised_time = []
for n in mat_size:
    start = time.time()
    matrix1 = np.random.randint(200,size = (n,n))
    matrix2 = np.random.randint(200,size = (n,n))
    
    res = np.dot(matrix1,matrix2)
    end = time.time()
    time_taken = end-start
    vectorised_time.append(time_taken)
```

```python
# Creaing a dataframe

pd.DataFrame({"matrix size":mat_size,
              "non_vectorised_time":non_vectorised_time,
             "vectorised_time":vectorised_time})
```

```python
# Plotting vectorised and non vectorised execution time

plt.plot(mat_size,non_vectorised_time, label="Non Vectorised method")
plt.plot(mat_size,vectorised_time, label="Vectorised method" )
plt.xlabel("matrix size")
plt.ylabel("Time taken")
plt.title("Comparision of time required for vectorised and non vectorised methods")
plt.legend()
plt.show()
```
